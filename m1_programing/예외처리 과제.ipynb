```python
import time
import random
print('='*50)
print('오늘도 점심이 걱정이신가요? 점심을 추천해 드릴게요!')
print('='*50)
time.sleep(1)
rice = ['김치볶음밥','제육덮밥','낙지덮밥','생선구이정식','삼겹살정식']
noodle = ['라면','우동','쫄면','냉면','파스타','짜장면','짬뽕']
soup = ['돼지국밥','순두부찌개','김치찌개','된장찌개']
while 1:
    try:
        pick = int(input('밥이냐 면이냐 국이냐 그대의 선택은~!? 밥:1 면:2 국:3 종료:0' ))
        if pick > 4 or pick < 0:
            raise       
    except ValueError:
        print('문자가 아닌 숫자로 부탁드릴게요!')
    except:
        print('보기의 숫자로 부탁드릴게요!')
    if 0 < pick or pick > 3:
        if pick == 1:
            try:
                print('오늘의 메뉴는 바로바로~!!' + random.choice(rice))
                like = int(input('''메뉴가 마음에 든다면 1,
다시 고르고 싶다면 2,
도움이 안된다면 0을 누르세요ㅠ_ㅠ'''))
                if like > 3 or like <0:
                    raise
            except ValueError:
                print('문자가 아닌 숫자로 부탁드릴게요!')
            except:
                print('보기의 숫자로 부탁드릴게요!')
            else:
                if like == 0:
                    print('혼자서도 잘할 수 있죠?! 다음엔 마음에 드는 픽을 가져올게요!')
                    break
                elif like == 1:
                    print('오늘도 즐겁고 맛있는 점심 식사가 되시길~!')
                    break
                elif like == 2:
                    print(' 맘에 안드신다면 다시 골라 드릴게요!')
        elif pick == 2:
            try:
                print('오늘의 메뉴는 바로바로~!!' + random.choice(noodle))
                like = int(input('''메뉴가 마음에 든다면 1,
다시 고르고 싶다면 2,
도움이 안된다면 0을 누르세요ㅠ_ㅠ'''))
                if like > 3 or like <0:
                    raise
            except ValueError:
                print('문자가 아닌 숫자로 부탁드릴게요!')
            except:
                print('보기의 숫자로 부탁드릴게요!')
            else:
                if like == 0:
                    print('혼자서도 잘할 수 있죠?! 다음엔 마음에 드는 픽을 가져올게요!')
                    break
                elif like == 1:
                    print('오늘도 즐겁고 맛있는 점심 식사가 되시길~!')
                    break
                elif like == 2:
                    print(' 맘에 안드신다면 다시 골라 드릴게요!')
        elif pick == 3:
            try:
                print('오늘의 메뉴는 바로바로~!!' + random.choice(soup))
                like = int(input('''메뉴가 마음에 든다면 1,
다시 고르고 싶다면 2,
도움이 안된다면 0을 누르세요ㅠ_ㅠ'''))
                if like > 3 or like <0:
                    raise
            except ValueError:
                print('문자가 아닌 숫자로 부탁드릴게요!')
            except:
                print('보기의 숫자로 부탁드릴게요!')
            else:
                if like == 0:
                    print('혼자서도 잘할 수 있죠?! 다음엔 마음에 드는 픽을 가져올게요!')
                    break
                elif like == 1:
                    print('오늘도 즐겁고 맛있는 점심 식사가 되시길~!')
                    break
                elif like == 2:
                    print(' 맘에 안드신다면 다시 골라 드릴게요!')
    else:
        print('점심에 대한 고민이 없어지는 그날까지~! 다음에 다시 만나요')
        break
```
